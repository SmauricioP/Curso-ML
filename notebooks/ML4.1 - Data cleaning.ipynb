# Carga de datos a través de la función open

Aunque no esté pensado para ello, siempre se puede hacer la carga de un fichero a mano. A la hora de trabajar con datasets muy grandes, pandas no es óptima y se demora mucho en funcionar, comiéndose la memorias del ordenador (¡se puede tener hasta ficheros de 15GB!).
Se puede utilizar uno de los métodos de ficheros que lee dataset linea por linea: open.

```python
df3 = open(mainpath + "customer-churn-model/Customer Churn Model.txt", 'r') # r es modo lectura
```

Readline lee toda la línea y lo pone en un string. El método strip elimina espacios en blanco que sobran al inicio o final de la línea. Split divide en listas el string dependiendo del delimitador.

```python
firstline = df3.readline()
```

```python
cols = firstline.strip().split(",")
n_cols = len(cols) # Longitud
cols
```

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
for line in df3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter +=1
```

```python
print("El data set tiene %d filas y %d columnas" %(counter, n_cols))
```

Ahora, el main.dict tiene toda la información relevante sobre el dataset, por lo que se le puede convertir en un dataframe.

```python
data3 = pd.DataFrame(main_dict)
data3.head()
```

Se debe tener cuidado con el método open al abrir ficheros en modo "w" (escritura), pues esto hará que dicho fichero pierda su contenido. Es recomendable agregar contenido en modo "a" (append).

Algo típico es que el delimitador sea un "barra t". Se suele usar como delimitador cuando se exporta una base de datos.

## Lectura y escritura de ficheros

```python
infile  = mainpath + "customer-churn-model/Customer Churn Model.txt"
outfile = mainpath + "customer-churn-model/Tab Customer Churn Model.txt"
```

```python
with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write('\t'.join(fields))
            outfile1.write('\n') # Intro
```

Así no exista el fichero inicialmente, Python se encarga de crearlo.

```python
df4 = pd.read_csv(outfile, sep = "\t")
df4.head()
```

# Leer datos desde una URL

```python
medals_url = 'https://raw.githubusercontent.com/joanby/python-ml-course/master/datasets/athletes/downloaded_medals.csv'
iris_url = 'https://raw.githubusercontent.com/joanby/python-ml-course/master/datasets/iris/iris.csv'
```

```python
df_medals = pd.read_csv(medals_url)
df_medals = df_medals.drop(df_medals.columns[[0]],1) # Borra columnas
```

```python
import csv
import urllib3
```

Urllib3 navega y obtiene información desde una URL. El paquete csv realiza un mejor manejo sobre csv.

El siguiente chunk es una función que permite obtener datos directamente de una URL. Se puede especificar el separador, delimitador, el formato para decodificar y, obviamente, tanto el url como el nombre del dataset.

```python
def FromUrl(data_name, url, sep = ",", delim = "\n", encoding = "utf-8"):
    # Sacar datos de una URL y decodificación
    http = urllib3.PoolManager()
    r = http.request('GET',url)
    r.status
    response = r.data.decode(encoding).strip().split(delim) # Tipo byte

    # Nombre de columnas
    cols_medals = response[0].split(sep)[1:]

    # Utiizar reader y construir el dataframe
    cr = csv.reader(response)
    df_test = pd.DataFrame(cr)
    df_test = df_test.drop(df_test.columns[[0]],1).drop(0,0)
    df_test.columns = cols_medals

    # Número de columnas 
    print("El dataset {dataname} tiene {ncols} features y {nrows} observaciones".format(
        dataname = data_name,
        ncols    = len(cols_medals),
        nrows    = len(df_test))
         )
    
    # Asignar el nombre y la ruta para escribir un nuevo archivo
    filename = "dummy/downloaded_" + data_name +"."
    fullpath = os.path.join(mainpath,filename)

    # Escribir el archivo en tres formatos distintos
    print('Guardando fichero en csv, json y xlsx...')
    df_test.to_csv(fullpath + "csv")
    df_test.to_json(fullpath + "json")
    df_test.to_excel(fullpath + "xlsx")
    print('Los ficheros se han guardado correctamente en:',
         fullpath)
    
    if data_name in ['iris']:
        print('Nota: El dataset',data_name,'es utilizado frecuentemente.')
    # Head del dataset
    return df_test.head()
```

```python
FromUrl('iris', iris_url, sep =",")
```
