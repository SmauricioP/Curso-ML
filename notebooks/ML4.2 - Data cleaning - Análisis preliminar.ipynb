```python
pd.DataFrame(pd.isnull(data['body'])).describe()
```

```python
pd.notnull(data['body']).values.ravel().sum()  # Ravel convierte en un solo array
```

Los valores que faltan en un data set pueden venir por dos razones:
* Extracción de los datos
* Recolección de los datos

```python
def createDummies(df, var_name, write = False):
    dummy = pd.get_dummies(df[var_name], prefix = var_name)
    df = df.drop(var_name, axis = 1)
    if write == True:
        df = pd.concat([df, dummy], axis = 1)
        print('Dataset sobreescrito con las dummies.')
        return df
    else:
        print('No se ha sobrescrito el dataset.',
              'De sobreescribirse, el nuevo dataset sería:')
        return pd.concat([df, dummy], axis = 1).head(10)
    
    
```
